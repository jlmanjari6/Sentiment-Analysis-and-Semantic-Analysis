```python
# looping through all the 1578 files in the articles folder to use the search query “Canada”, “Halifax”, “Nova Scotia”, 
# and search in how many documents these words have appeared.

import pandas as pd
import math
import os
    
path = './articles/'

canada_count = 0
halifax_count = 0
ns_count = 0

df_canada = pd.DataFrame(columns=['Canada appeared in documents', 'Total Words (m)', 
                                  'Frequency (f)','Relative frequency(f/m)'])
article_count = 0

for file in os.listdir(path):
    canada_count_in_article = 0
    with open(os.path.join('./articles/',file)) as fp:        
        content = fp.read()        
        total_words = content.split()        
        list_words_file = re.split('(?<!nova)\s+', content)
        list_words_file_lowercase = [x.lower() for x in list_words_file]        
        if 'canada' in list_words_file_lowercase:
            canada_count += 1            
        if 'halifax' in list_words_file_lowercase:
            halifax_count += 1
        if 'nova scotia' in list_words_file_lowercase:
            ns_count += 1
               
        for word in list_words_file_lowercase:
            if word == 'canada':
                canada_count_in_article += 1
                
        fp.close()
    df_canada.loc[article_count] = [file, len(total_words), canada_count_in_article, canada_count_in_article/len(total_words)]
    article_count += 1
        

df = pd.DataFrame(columns=['Search Query', 'Document containing term(df)', 
                           'Total Documents(N)/ number of documents term appeared (df)', 'Log10(N/df)'])

df.loc[0] = ['Canada', canada_count, '1578/' + str(canada_count), math.log((1578/canada_count), 10)]
df.loc[1] = ['Halifax', halifax_count, '1578/' + str(halifax_count), 'NA']
df.loc[2] = ['Nova Scotia', ns_count, '1578/' + str(ns_count), 'NA']
```
